```python
def average1(S):
    n = len(S)
    my_average = [0] * n
    for j in range(n):
        total = 0
    for i in range(j + 1):
        total += S[i]
    my_average[j] = total / (j+1)
    return my_average
```

```python
def average2(S):
  #S:sequence
    n = len(S)
    my_average= [0] * n
    for j in range(n):
        my_average[j] = sum(S[0:j+1]) / (j+1)
    return my_average
```

```python
def average3(S):
 #S:sequence
    n = len(S)
    my_average = [0] * n
    total = 0
    for j in range(n):
        total += S[j]
        my_average[j] = total / (j+1)
    return my_average
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
algo1 = []
algo2 = []
algo3 = []
from timeit import default_timer
for i in range(100,5000,100):
    input_arr = list(range(i))
    start = default_timer()
    average1(input_arr)
    end = default_timer()
    algo1.append([i,end-start])
    
    start = default_timer()
    average2(input_arr)
    end = default_timer()
    algo2.append([i,end-start])
    
    start = default_timer()
    average3(input_arr)
    end = default_timer()
    algo3.append([i,end-start])
```

```python
algo1 = np.array(algo1)
algo2 = np.array(algo2)
algo3 = np.array(algo3)
```

```python
plt.figure(figsize=(10, 10), dpi=70)
plt.loglog(algo1[:,0], algo1[:,1],  label = "Algorithm 1")
plt.loglog(algo2[:,0], algo2[:,1],  label = "Algorithm 2")
plt.loglog(algo3[:,0], algo3[:,1],  label = "Algorithm 3")

plt.ylabel("Time")
plt.xlabel("Size")
plt.legend()
```
